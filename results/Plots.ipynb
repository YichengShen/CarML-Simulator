```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cifar_cgc_sign_flip_acc = np.zeros(29)
cifar_cgc_sign_flip_loss = np.zeros(29)

for i in range(1, 11):
    f = f'cifar10-cgc-signflip-round{i}.csv'
    df = pd.read_csv(f, header=None)
    cifar_cgc_sign_flip_acc = np.add(cifar_cgc_sign_flip_acc, np.array(df[1]))
    cifar_cgc_sign_flip_loss = np.add(cifar_cgc_sign_flip_loss, np.array(df[2]))

cifar_cgc_sign_flip_acc = np.true_divide(cifar_cgc_sign_flip_acc, 10)
cifar_cgc_sign_flip_loss = np.true_divide(cifar_cgc_sign_flip_loss, 10)

print(cifar_cgc_sign_flip_acc)
print(cifar_cgc_sign_flip_loss)
```

```python
cifar_mean_sign_flip_acc = np.zeros(29)
cifar_mean_sign_flip_loss = np.zeros(29)

for i in range(1, 3):
    f = f'cifar10-simplemean-signflip-round{i}.csv'
    df = pd.read_csv(f, header=None)
    cifar_mean_sign_flip_acc = np.add(cifar_mean_sign_flip_acc, np.array(df[1]))
    cifar_mean_sign_flip_loss = np.add(cifar_mean_sign_flip_loss, np.array(df[2]))

cifar_mean_sign_flip_acc = np.true_divide(cifar_mean_sign_flip_acc, 2)
cifar_mean_sign_flip_loss = np.true_divide(cifar_mean_sign_flip_loss, 2)

print(cifar_mean_sign_flip_acc)
print(cifar_mean_sign_flip_loss)
```

```python
cifar_mean_noattack_acc = np.zeros(29)
cifar_mean_noattack_loss = np.zeros(29)

df = None

for i in range(1, 11):
    f = f'cifar10-simplemean-noattack-round{i}.csv'
    df = pd.read_csv(f, header=None)
    cifar_mean_noattack_acc = np.add(cifar_mean_noattack_acc, np.array(df[1]))
    cifar_mean_noattack_loss = np.add(cifar_mean_noattack_loss, np.array(df[2]))

cifar_mean_noattack_acc = np.true_divide(cifar_mean_noattack_acc, 10)
cifar_mean_noattack_loss = np.true_divide(cifar_mean_noattack_loss, 10)

print(cifar_mean_noattack_acc)
print(cifar_mean_noattack_loss)
```

```python
x_axis = np.array(df[0])
print(x_axis)
```

```python
plt.plot(x_axis, cifar_mean_noattack_acc, c='dodgerblue')
plt.plot(x_axis[9], cifar_mean_noattack_acc[9], '-o', label="MEAN without failures", c='dodgerblue')
plt.plot(x_axis, cifar_cgc_sign_flip_acc, c='black', linestyle='dashed')
plt.plot(x_axis[10], cifar_cgc_sign_flip_acc[10], 'D', label="CRACAU", c='black', linestyle='dashed')
plt.plot(x_axis, cifar_mean_sign_flip_weak_acc, c='orange', linestyle='dotted')
plt.plot(x_axis[8], cifar_mean_sign_flip_weak_acc[8], 'v', label="MEAN", c='orange', linestyle='dotted')


plt.xscale('log')
plt.ylim(bottom=0, top=0.9)
plt.yticks(np.arange(0, 0.9, 0.2))
plt.legend()
plt.xlabel('Epoch')
plt.ylabel("Top-1 accuracy")

plt.show()
```

```python
plt.plot(x_axis, cifar_mean_noattack_loss, c='dodgerblue')
plt.plot(x_axis[9], cifar_mean_noattack_loss[9], '-o', label="MEAN without failures", c='dodgerblue')
plt.plot(x_axis, cifar_cgc_sign_flip_loss, c='black', linestyle='dashed')
plt.plot(x_axis[10], cifar_cgc_sign_flip_loss[10], 'D', label="CRACAU", c='black', linestyle='dashed')
plt.plot(x_axis, cifar_mean_sign_flip_weak_loss, c='orange', linestyle='dotted')
plt.plot(x_axis[8], cifar_mean_sign_flip_weak_loss[8], 'v', label="MEAN", c='orange', linestyle='dotted')


plt.xscale('log')
plt.ylim(bottom=0, top=2.4)
plt.yticks(np.arange(2.0, -0.1, -0.5))
plt.legend()
plt.xlabel('Epoch')
plt.ylabel("Loss")

plt.show()
```
